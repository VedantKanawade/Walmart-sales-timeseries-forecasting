```python
import pandas as pd
ts_df = pd.read_csv("../data/processed/ts_stationary.csv", parse_dates=['Date'], index_col='Date')
```

```python
for lag in [1,2,3,4]:
    ts_df[f'lag_{lag}'] = ts_df['Weekly_Sales'].shift(lag)
```

```python
ts_df['rolling_mean_4'] = ts_df['Weekly_Sales'].rolling(window=4).mean()
ts_df['rolling_std_4'] = ts_df['Weekly_Sales'].rolling(window=4).std()
```

```python
ts_df['rolling_mean_12'] = ts_df['Weekly_Sales'].rolling(window=12).mean()
ts_df['rolling_std_12'] = ts_df['Weekly_Sales'].rolling(window=12).std()
```

```python
ts_df['week_of_year'] = ts_df.index.isocalendar().week
ts_df['month'] = ts_df.index.month
ts_df['quarter'] = ts_df.index.quarter
ts_df['year'] = ts_df.index.year
```

```python
features = pd.read_csv("../data/processed/features_typed.csv")
features['Date'] = pd.to_datetime(features['Date'])

```

```python
holidays = features[['Date','IsHoliday']].drop_duplicates()
```

```python
print(holidays.dtypes)
```

```python
ts_df = ts_df.reset_index()  # Date becomes a column
```

```python
ts_df = ts_df.merge(holidays, on='Date', how='left')
```

```python
ts_df['IsHoliday'] = ts_df['IsHoliday'].fillna(False)
```

```python
ts_df.set_index('Date', inplace=True)
```

```python
ts_df.to_csv("../data/processed/ts_features.csv")
```
